```python
import os
import urllib.request as req
import subprocess
from imghdr import what
from PIL import Image
```

```python
def wget(link, fname, t):
    subprocess.call(['wget', link, '-O', fname], timeout=t)
    if what(fname) == None:
        subprocess.call(['rm', fname], timeout=t)
        raise Exception("file corrupted")
    h,w = Image.open(fname).size
    if h < 224 or w < 224:
        subprocess.call(['rm', fname], timeout=t)
        raise Exception("file corrupted or small")
```

```python
with open("/home/lestsovba/dataset/imagenet/valid_labels.txt") as f:
    lines = f.readlines()
    labels = [i.split() for i in lines]
    valids = {i[0]:int(i[1]) for i in labels}
```

```python
NUMFILES = 100
TIMEOUT = 3

fname = "/home/lestsovba/dataset/imagenet/list.txt"
dest = "/home/lestsovba/dataset/imagenet/images/"

i = 0
with open(fname, 'r') as f, open("res.txt", 'w') as o:
    while i < NUMFILES:
        line = f.readline().split(maxsplit=1)
        link = line[1]
        ident = line[0][:9]
        
        lblname = labels[valids[ident]][2]
        name = link.split('/')[-1]
        dpath = os.path.join(dest, name)
        print("Getting", lblname, name, link, end=' ', flush=True)

        try:
            wget(link, dpath, TIMEOUT)
        except Exception as e:
            print("err:", e)
        else:
            print("ok")
            i += 1
```
